```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
races_df=pd.read_csv("Data/races.csv")
races_df.head()
```

```python
runs_df=pd.read_csv("Data/runs.csv")
runs_df.head()
```

```python
horse_df = pd.read_csv("Data/race-result-horse.csv")
horse_df = horse_df[["finishing_position","horse_name","jockey","trainer","actual_weight","declared_horse_weight","finish_time","race_id"]]
horse_df.head()
```

```python
race_df= pd.read_csv("Data/race-result-race.csv")
race_df=race_df[["race_id","race_class","race_distance"]]
race_df.head()
```

```python
merge_df=pd.merge(race_df,horse_df,on="race_id",how="outer")
merge_df.head()
```

```python
merge_df["finishing_position"].value_counts()
```
